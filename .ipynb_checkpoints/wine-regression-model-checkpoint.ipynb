# Red and White Wine Linear Regression Models

The purpose of this project is to build two regression models (red/white) 
that will predict the rating of a selected wine based on select measurements, 
and also try to develop a model that can comprehend both types of wine. 

```python
#importing libraries

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import pylab
from keras.models import load_model
```

## EDA

Let's begin by looking at each of the datasets to find correlations 

```python
red_wine_label = "winequality-red.csv"
white_wine_label = "winequality-white.csv"

red_dir = os.path.join("data", red_wine_label)
white_dir = os.path.join("data", white_wine_label)

red = pd.read_csv(red_dir, delimiter = ";")
white = pd.read_csv(white_dir, delimiter = ";")
```

```python
print('Length of red wine data:', len(red))
print('Length of white wine data:', len(white))
```

```python
red_combined = red.copy()
red_combined['type'] = 'red'

white_combined = white.copy()
white_combined['type'] = 'white'

combined = pd.concat([red_combined, white_combined])
combined.head()
```

```python
combined.shape
```

```python
red.info()
white.info()
```

```python
combined = combined.reset_index(drop=True)
```

```python
wine_displot = sns.displot(x = 'quality', 
            data = combined, 
            hue = "type", 
            height = 12, 
            palette = ["#9b0000", "#8B8000"])
```

```python
fig = stats.probplot(combined['quality'], dist = 'norm', plot = pylab)
pylab.show()
```

We can see that the most common quality score is a 6, and through the qqplot we can also confirm that the provided data is normally distributed

```python
plt.figure(figsize=(18,12))
plt.title('White Wine Correlation')
sns.heatmap(red.corr(), cmap = "flare", annot = True)
```

```python
plt.figure(figsize=(18,12))
plt.title('White Wine Correlation')
sns.heatmap(white.corr(), cmap = "crest", annot = True)
```

Observing the correlation plots, we can see that both red and white wine's scores are heavily affected by its alcohol content (0.44-0.48). 

Let's see the most effective qualities:
- **Sulphates**: Red wine has a correlation value of 0.25 for sulphate, whereas white wine has a correlation score of 0.054, which means red wine is more valued for its depth of flavor, where white wine is valued for its lighter flavors 
- **pH**: Red wine has a _negative_ correlation to pH, meaning its value increases when it becomes more bitter, whereas white wine has a _positive_ correlation to pH, meaning its value increases as it becomes more acidic/sour. 
- **Citric Acid**: Inversely, red wine has a negative correlation value to quality whereas white wine has a positive value

We can remove any attributes that are deemed not significant -- we can remove free sulfur dioxide because its correlation value is the smallest on average 

```python
red.drop('free sulfur dioxide', axis = 1)
white.drop('free sulfur dioxide', axis = 1)
```

Let's explore the acidity further:

```python
sns.displot(x='pH',
            y='quality',
            data=combined, 
            hue = 'type', 
            height = 10, 
            palette = ["#9b0000", "#8B8000"])
```

As you can see, there is a notable divide between the red and white wine in terms of pH, but the rating is too sparse to conclude whether it has an effect.
